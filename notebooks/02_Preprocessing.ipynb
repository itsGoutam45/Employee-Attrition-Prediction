```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../data/raw_data/WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
df_copy = df.copy()
```

```python
df_copy.columns
```

```python
df_copy.shape
```

```python
df_copy.drop(columns=["EmployeeCount", "StandardHours", "EmployeeNumber"], inplace=True)
#EmployeeNumber is just an ID
```

```python
df_copy.columns
```

```python
df_copy.shape
```

```python
numeric_columns = df_copy.select_dtypes(include=["int","float"]).columns
```

```python
catagorical_columns = df_copy.select_dtypes(include=["object"]).columns
```

```python
X = df_copy.drop("Attrition", axis=1)
```

```python
X.head()
```

```python
y = df_copy["Attrition"]
```

```python
y.head()
```

```python
y = y.map({"Yes":1, "No":0})
```

```python
y.head()
```

```python
X.head()
```

```python
X.shape
```

```python
X = pd.get_dummies(X, drop_first=True, dtype=int)
```

```python
X.shape
```

```python
X.head()
```

```python
X.info()
```

##### We'll going  to scall only the continuous numeric columns...

```python
binary_columns = []
for col in X.columns:
    if X[col].nunique()==2:
        binary_columns.append(col)
```

```python
print(len(binary_columns))
binary_columns
```

```python
continuous_columns = []
for col in X.columns:
    if col not in binary_columns:
        continuous_columns.append(col)
```

```python
print(continuous_columns, len(continuous_columns), sep="\n")
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
print(y.value_counts(normalize=True))

print(y_train.value_counts(normalize=True))

print(y_test.value_counts(normalize=True))

#the ratio is preserved because of stratify, and we should use stratify here because the dataset is imbalanced
```

### Scalling

```python
scaler = StandardScaler()
```

```python
x_train[continuous_columns] = scaler.fit_transform(x_train[continuous_columns])
```

```python
x_train.head()
```

```python
x_test[continuous_columns] = scaler.transform(x_test[continuous_columns])
```

```python
x_test.head()
```

```python
x_train[continuous_columns].describe()
```

```python
x_train.to_csv("../data/processed_data/x_train.csv", index=False)
x_test.to_csv("../data/processed_data/x_test.csv", index=False)
y_train.to_csv("../data/processed_data/y_train.csv", index=False)
y_test.to_csv("../data/processed_data/y_test.csv", index=False)
```
